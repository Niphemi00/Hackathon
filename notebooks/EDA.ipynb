```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
math_data = pd.read_csv('../data/student_mat.csv')
por_data = pd.read_csv('../data/student_por.csv')
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
print(f"Math dataset shape: {math_data.shape}")
print(f"Portuguese dataset shape: {por_data.shape}")

# Datasets inspection
display(math_data.head())
display(por_data.head())

# Combine datasets
combined_student_df = pd.concat([math_data, por_data], ignore_index=True)
print(f"Combined dataset shape: {combined_student_df.shape}")

# Summary info
print("\n=== Dataset Info ===")
combined_student_df.info()

print("\n=== Missing Values ===")
print(combined_student_df.isnull().sum())

# Basic statistics
display(combined_student_df.describe())

# Distribution of final grade (G3)
sns.histplot(combined_student_df['G3'], bins=20, kde=True)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Count")
plt.show()

# Distribution of absences
sns.histplot(combined_student_df['absences'], bins=20, kde=True)
plt.title("Distribution of Absences")
plt.xlabel("Absences")
plt.ylabel("Count")
plt.show()

# Alcohol consumption
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.countplot(x="Dalc", data=combined_student_df, ax=axes[0], palette="viridis")
axes[0].set_title("Workday Alcohol Consumption (1=low, 5=high)")

sns.countplot(x="Walc", data=combined_student_df, ax=axes[1], palette="mako")
axes[1].set_title("Weekend Alcohol Consumption (1=low, 5=high)")
plt.show()

# Gender vs. final grade
sns.boxplot(x="sex", y="G3", data=combined_student_df)
plt.title("Final Grades by Gender")
plt.show()

# Save combined but still raw file
processed_path = "../data/combined_students_raw.csv"
combined_student_df.to_csv(processed_path, index=False)
```
